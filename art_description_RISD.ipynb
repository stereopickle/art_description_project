### Import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import re
%matplotlib inline
```

```python
url = "https://risdmuseum.org/api/v1/collection" #RISD Museum collection
```

```python
# This API allow loading maximum 25 items at a time. 
# There are about 3900 works with 'painting' search term in their collection.
# So run it through about 156 times.
db = []
for i in range(156):
    resp = requests.get(url, {'search_api_fulltext': 'painting', 
                'items_per_page': 25, 
                'page': i})
    db.extend(resp.json())
```

```python
df = pd.DataFrame(db)
len(df)
```

What are some of the most commonly used words to describe paintings for each period of time? 
What are some of the most commonly used words to describe paintings for each artists?
What are the words that describe art when the word 'abstract' is used in the work vs. when it's not.



```python
# only select items with some description.
desc_df = df[df['description'] != '']
desc_df.description = desc_df.description.apply(lambda x: re.sub('[^A-Za-z0-9]+', ' ', x).lower())
```

```python
word_list = ' '.join(list(desc_df.description)).split()
```

```python
unique_words = set(word_list)
```

```python
word_counts = dict.fromkeys(unique_words, 0)
```

```python
for word in word_list: 
    word_counts[word] += 1
```

```python
{k: v for k, v in sorted(word_counts.items(), reverse = True, key=lambda item: item[1])}
```
